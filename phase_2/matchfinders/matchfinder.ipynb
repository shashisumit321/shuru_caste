```python
import pandas as pd
```

```python
mf = pd.read_excel("CastesAndSubcastesInIndia.xlsx",header=1)
```

```python
mf
```

```python
mf["State Name"].replace({
    "GUJA RAT": "GUJARAT",
    "HARYA NA": "HARYANA",
    "HIMACHA L PRADESH": "HIMACHAL PRADESH"
}, inplace=True)
```

```python
import pandas as pd

# Group by state and count unique non-null values
result_df = mf.groupby('State Name').agg(
    unique_castes=('Caste', lambda x: x.dropna().nunique()),
    unique_subcastes=('Subcaste', lambda x: x.dropna().nunique())
).reset_index()

# Save to CSV
# result_df.to_csv('state_caste_subcaste_counts.csv', index=False)

# Display the DataFrame
result_df
```

```python
result_df.to_csv('no_of_castes_and_subcastes_state_wise_in_matchfinder.csv', index=False)
```
