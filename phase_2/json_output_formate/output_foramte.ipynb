```python
import json
import os
from multiprocessing import Pool, cpu_count
from googletrans import Translator

INPUT_FILE = "output/caste_structure_base.json"
OUTPUT_FILE = "output/caste_structure_final_transliterated.json"

with open(INPUT_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)
    data = data[:5]  # For testing, limit to 2 entries

LANGUAGES = ["en", "hi", "as", "bn", "gu", "kn", "ml", "mr", "or", "pa", "ta", "te", "or"]  # 'ori' -> 'or'

def translate_entry(entry):
    translator = Translator()
    name = entry['name']
    display_name = {}
    for lang in LANGUAGES:
        try:
            translated = translator.translate(name, src='en', dest=lang)
            display_name[lang] = translated.text
        except Exception:
            display_name[lang] = name
    entry['displayName'] = display_name
    return entry

if __name__ == "__main__":
    os.makedirs("output", exist_ok=True)
    with Pool(max(cpu_count() - 1, 2)) as pool:
        results = pool.map(translate_entry, data)

    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        json.dump(results, f, indent=2, ensure_ascii=False)

    print(f"✅ Final transliterated JSON saved to: {OUTPUT_FILE}")
```

```python
from googletrans import Translator
from concurrent.futures import ThreadPoolExecutor, as_completed

translator = Translator()
langs = ["hi", "bn", "gu", "kn", "ml", "mr", "or", "pa", "ta", "te", "as"]
text = "lohar"

def translate_text(lang):
    try:
        translated = translator.translate(text, src='en', dest=lang)
        return lang, translated.text
    except Exception as e:
        return lang, f"❌ Error: {str(e)}"

results = {}

with ThreadPoolExecutor(max_workers=5) as executor:
    future_to_lang = {executor.submit(translate_text, lang): lang for lang in langs}
    for future in as_completed(future_to_lang):
        lang, result = future.result()
        results[lang] = result

# Print results
for lang, translated in results.items():
    print(f"{lang}: {translated}")
```

```python
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate

# Your defined language to script mapping
lang_map = {
    "en": sanscript.ITRANS,  # English often uses ITRANS for input
    "hi": sanscript.DEVANAGARI,    # Hindi
    "as": sanscript.BENGALI,       # Assamese uses Bengali script
    "bn": sanscript.BENGALI,
    "gu": sanscript.GUJARATI,
    "kn": sanscript.KANNADA,
    "ml": sanscript.MALAYALAM,
    "mr": sanscript.DEVANAGARI,    # Marathi
    "or": sanscript.ORIYA,
    "pa": sanscript.GURMUKHI,
    "ta": sanscript.TAMIL,
    "te": sanscript.TELUGU,
}

text_to_transliterate = "lohar" # The word you want to transliterate

print(f"Transliterating '{text_to_transliterate}':\n")

# Assuming the input "mix" is in a basic Roman script that can be mapped to ITRANS
# We'll start from ITRANS as the source for simplicity
source_script = sanscript.ITRANS

for lang_code, target_script_enum in lang_map.items():
    if lang_code == "en": # Skip English as it's typically the source or a Romanization scheme
        continue

    transliterated_text = transliterate(
        text_to_transliterate,
        source_script,
        target_script_enum
    )
    print(f"  {lang_code.upper()} ({target_script_enum.title()}): {transliterated_text}")
```
