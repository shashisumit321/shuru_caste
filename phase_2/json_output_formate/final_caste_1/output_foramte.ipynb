```python
import pandas as pd
import uuid
import json
import os
import ast

# --- Config ---
OUTPUT_DIR = "output"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "caste_structure_final_with_uid.json")
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Load data
df = pd.read_csv("final_caste.csv")
df.rename(columns={"subcastes": "Subcastes"}, inplace=True)

# --- Helper Functions ---
def normalize(text):
    return str(text).strip().replace(" ", "-").lower() if pd.notna(text) else ""

def clean_name(text):
    return str(text).strip().strip("'").strip('"')

def generate_display_name(name):
    name = clean_name(name)
    return {lang.capitalize(): name for lang in [
        "en", "hi", "as", "bn", "gu", "kn", "ml", "mr",
        "or", "pa", "ta", "te", "ori"
    ]}

def safe_eval(val):
    try:
        val = str(val).strip()
        if val.startswith("[") and val.endswith("]"):
            return [clean_name(x) for x in ast.literal_eval(val)]
        else:
            return [clean_name(x) for x in val.split(",") if clean_name(x)]
    except Exception:
        return []

# --- Step 1: Clean and Prepare Subcastes ---
df['Caste'] = df['Caste'].astype(str).apply(clean_name)
df['Subcastes'] = df['Subcastes'].fillna("[]").apply(safe_eval)

# --- Step 2: Generate UUIDs ---
caste_uid_map = {c: str(uuid.uuid4()) for c in df['Caste'].unique()}
df['CasteUID'] = df['Caste'].map(caste_uid_map)

# --- Step 3: Caste Records ---
caste_records = []
for caste, uid in caste_uid_map.items():
    caste_records.append({
        "id": uid,
        "name": caste,
        "displayName": generate_display_name(caste),
        "casteType": "CASTE",
        "key": normalize(caste),
        "parentKey": None,
        "parentId": None,
        "stateId": None
    })

# --- Step 4: Subcaste Records ---
subcaste_records = []
for _, row in df.iterrows():
    caste = row['Caste']
    parent_id = row['CasteUID']
    for sub in row['Subcastes']:
        subcaste_records.append({
            "id": str(uuid.uuid4()),
            "name": sub,
            "displayName": generate_display_name(sub),
            "casteType": "SUB_CASTE",
            "key": f"{normalize(caste)}#{normalize(sub)}",
            "parentKey": normalize(caste),
            "parentId": parent_id,
            "stateId": None
        })

# --- Final Output ---
final_json = caste_records + subcaste_records

with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(final_json, f, indent=2, ensure_ascii=False)

print(f"✅ JSON saved to: {OUTPUT_FILE}")
```

```python
import pandas as pd
import uuid
import json
import os
import ast

# --- Config ---
OUTPUT_DIR = "output"
OUTPUT_FILE = os.path.join(OUTPUT_DIR, "caste_structure_base.json")
os.makedirs(OUTPUT_DIR, exist_ok=True)

# --- Helpers ---
def normalize(text):
    return str(text).strip().replace(" ", "-").lower() if pd.notna(text) else ""

def clean_name(text):
    return str(text).strip().strip("'").strip('"')

def safe_eval(val):
    try:
        val = str(val).strip()
        if val.startswith("[") and val.endswith("]"):
            return [clean_name(x) for x in ast.literal_eval(val)]
        else:
            return [clean_name(x) for x in val.split(",") if clean_name(x)]
    except Exception:
        return []

# --- Load and Clean ---
df = pd.read_csv("final_caste.csv")
df.rename(columns={"subcastes": "Subcastes"}, inplace=True)

df['Caste'] = df['Caste'].astype(str).apply(clean_name)
df['Subcastes'] = df['Subcastes'].fillna("[]").apply(safe_eval)
df['source'] = df['source'].fillna('').astype(str)

# --- UUID Map ---
caste_uid_map = {c: str(uuid.uuid4()) for c in df['Caste'].unique()}
df['CasteUID'] = df['Caste'].map(caste_uid_map)

# --- Caste Records ---
caste_records = []
for _, row in df.drop_duplicates(subset='Caste').iterrows():
    caste = row['Caste']
    caste_records.append({
        "id": caste_uid_map[caste],
        "name": caste,
        "displayName": {},  # will be filled later
        "casteType": "CASTE",
        "key": normalize(caste),
        "parentKey": None,
        "parentId": None,
        "stateId": None,
        "source": row['source']
    })

# --- Subcaste Records ---
subcaste_records = []
for _, row in df.iterrows():
    caste = row['Caste']
    parent_id = row['CasteUID']
    source = row['source']
    for sub in row['Subcastes']:
        subcaste_records.append({
            "id": str(uuid.uuid4()),
            "name": sub,
            "displayName": {},  # will be filled later
            "casteType": "SUB_CASTE",
            "key": f"{normalize(caste)}#{normalize(sub)}",
            "parentKey": normalize(caste),
            "parentId": parent_id,
            "stateId": None,
            "source": source
        })

# --- Combine and Save ---
final_json = caste_records + subcaste_records

with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(final_json, f, indent=2, ensure_ascii=False)

print(f"✅ Base JSON (no transliteration) saved to: {OUTPUT_FILE}")
```

```python
import json
import os
from multiprocessing import Pool, cpu_count
from googletrans import Translator

INPUT_FILE = "output/caste_structure_base.json"
OUTPUT_FILE = "output/caste_structure_final_transliterated.json"

with open(INPUT_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)
    data = data[:5]  # For testing, limit to 2 entries

LANGUAGES = ["en", "hi", "as", "bn", "gu", "kn", "ml", "mr", "or", "pa", "ta", "te", "or"]  # 'ori' -> 'or'

def translate_entry(entry):
    translator = Translator()
    name = entry['name']
    display_name = {}
    for lang in LANGUAGES:
        try:
            translated = translator.translate(name, src='en', dest=lang)
            display_name[lang] = translated.text
        except Exception:
            display_name[lang] = name
    entry['displayName'] = display_name
    return entry

if __name__ == "__main__":
    os.makedirs("output", exist_ok=True)
    with Pool(max(cpu_count() - 1, 2)) as pool:
        results = pool.map(translate_entry, data)

    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        json.dump(results, f, indent=2, ensure_ascii=False)

    print(f"✅ Final transliterated JSON saved to: {OUTPUT_FILE}")
```

```python
df['name'] = df['name'].astype(str).str.lower().str.replace(' ', '-')

# --- 2. Process the 'displayName' column (specifically 'En' key) ---
# We need a function to update the dictionary in each row
def standardize_display_name_en(display_dict):
    if isinstance(display_dict, dict) and 'En' in display_dict:
        display_dict['En'] = str(display_dict['En']).title() # Convert to Title Case
    return display_dict

# Apply the function to the 'displayName' column
df['displayName'] = df['displayName'].apply(standardize_display_name_en)
```

```python
df
```

```python
def convert_keys_to_lowercase(display_dict):
    if isinstance(display_dict, dict):
        return {key.lower(): value for key, value in display_dict.items()}
    return display_dict # Return as is if not a dictionary (e.g., NaN, None)

# Apply the function to the 'displayName' column
df['displayName'] = df['displayName'].apply(convert_keys_to_lowercase)
```

```python
import json

# Load the two files
with open("output/caste_structure_base.json", "r") as f1, open("output/caste_structure_final_transliterated_final.json", "r") as f2:
    data1_list = json.load(f1)
    data2_list = json.load(f2)

# Create a lookup from file1 by 'key'
source_lookup = {item["key"]: item["source"] for item in data1_list if "source" in item}

# Inject 'source' into file2 entries if matching key found
for item in data2_list:
    key = item.get("key")
    if key in source_lookup:
        item["source"] = source_lookup[key]
# Save updated file2
with open("output/file2_updated.json", "w", encoding="utf-8") as f_out:
    json.dump(data2_list, f_out, indent=2, ensure_ascii=False)
```
