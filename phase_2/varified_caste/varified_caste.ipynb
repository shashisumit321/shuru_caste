```python
gpt["source"] = "gpt"
gemini["source"] = "gemini"
manual_yes["source"] = "manual_yes"
manual_no["source"] = "manual_no"
```

```python
manual_no
```

```python
manual_caste_no = manual_no[["caste_name", "source"]].rename(columns={"caste_name": "Caste"})
```

```python
manual_caste_no
```

```python
gpt = gpt[['Caste', 'source']]
gemini = gemini[['Caste', 'source']]
manual_yes = manual_yes[['Caste', 'source']]
```

```python
all_caste = pd.concat([gpt, gemini, manual_yes, manual_caste_no], ignore_index=True)
```

```python
all_caste.dropna(inplace=True)
all_caste.drop_duplicates(inplace=True)
all_caste.reset_index(drop=True, inplace=True)
```

```python
all_caste
```

```python
all_caste.Caste = all_caste.Caste.str.strip().str.lower()
```

```python
all_caste.to_csv('all_varified_caste.csv', index=False)
```

```python
matchfinder = pd.read_excel('CastesAndSubcastesInIndia.xlsx', header=1)
```

```python
matchfinder
```

```python
matchfinder.Caste.nunique()
```

```python
matchfinder.Caste = matchfinder.Caste.str.strip().str.lower()
```

```python
import pandas as pd

# Sample merge (adjust column names if necessary)
result_df = matchfinder.merge(
    all_caste[['Caste', 'source']],  # only need caste and source from all_caste
    on='Caste',
    how='left'
)

# Add a new column 'type' based on whether caste was found in all_caste
result_df['validation_result'] = result_df['source'].apply(lambda x: 'Caste' if pd.notnull(x) else None)
```

```python
result_df
```

```python
varified_caste = result_df.dropna(subset=['validation_result'])
```

```python
result_df.drop_duplicates(subset=['State Name','Castecode','Caste','Subcaste'], inplace=True)
```

```python
result_df['State Name'] = result_df['State Name'].str.strip().str.lower()
result_df['Caste'] = result_df['Caste'].str.strip().str.lower()
result_df['Subcaste'] = result_df['Subcaste'].str.strip().str.lower()
result_df["source"] = result_df["source"].str.strip().str.lower()
result_df["validation_result"] = result_df["validation_result"].str.strip().str.lower()
```

```python
result_df.to_csv('matchfinder_caste_validation.csv', index=False)
```
