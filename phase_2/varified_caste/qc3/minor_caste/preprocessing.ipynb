```python
import pandas as pd
import ast

# Define the keywords for Muslim communities (all lowercase)
muslim_keywords = [
    "muslim", "saiyad", "syed", "sheikh", "qussab", "khan", "pathan",
    "ansari", "momin", "mullaji", "mughal", "naqvi", "rizvi", "sunni", "shia",
    "malik", "quraishi", "qureshi", "farooqui", "usmani", "peer", "fakir", "mufti"
]

# Function to check if any Muslim keyword is present
def contains_muslim_any(text):
    text = str(text).lower()
    return any(keyword in text for keyword in muslim_keywords)

# Convert Subcastes column from string to list
final_caste['subcastes'] = final_caste['subcastes'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])

# Function to check both caste and subcastes
def is_muslim_row(row):
    if contains_muslim_any(row['Caste']):
        return True
    return any(contains_muslim_any(sub) for sub in row['subcastes'])

# Apply filter
muslim_df = final_caste[final_caste.apply(is_muslim_row, axis=1)]
non_muslim_df = final_caste[~final_caste.apply(is_muslim_row, axis=1)]
```

```python
import pandas as pd
import ast

# Load the CSV file
df = pd.read_csv("muslim_caste.csv")
df["subcastes"] = df["subcastes"].apply(ast.literal_eval)

# Define Muslim caste list (standardized)
muslim_castes_raw = """
ABBASHI
ALI
ALVI
ANSAR
ARE KASAI
BAKSH
BALWAR
BEG
BEHNA
BHATIYAR
BHISTI
BOHRA
CHAMAR MUSLIM
CHIKWA QASSAB
DEKKANI MUSLIM
DHUNIA
DOGARWAR
DUDEKULA
DUDEKULA MANSURI
FAKEER
FAKIR
FAKIR BHANDARWALA
FAKIR INCLUDING ALVI-FAKIR
FAQIR
FAROOQ
GAARUDI
GADDI
GANCHI-PINJARA
GUJAR
GUJJAR(MUSLIM)
GUZAR
HALALKHOR
HASHMI
HUSSAIN
IDRISHI
IRAQI
JADOOGAR
JAMAN
JULAHA
JULAHA-ANSARI
JULAHA-MOMIN
JULAHE
JULAYA
KASAB
KASABI
KASAI-QURAISHI
KASSAB QUSSAB
KASSAR
KAZI
KHAKROB
KHAN
KHOTTA MUSLIM
LABBAI
LALBEG
LUHAR SAIFI
MADARI
MAKRANI
MAL MUSLIM
MALIK
MALKANA
MANERI
MANIHAR
MANIYAAR
MANSOORI
MANSUR
MANSURI-PINJARA
MAPPILLA
MEHTAR
MEKRANI
MIRZA
MIYAN
MOMIN
MOMIN-ANSARI
MUGHAL
MUJAHID
MUKATIB
MULLA
MUSKHAN
MUSLIM
MUSLIM BARUJIBI
MUSLIM BELDAR
MUSLIM BHANGI
MUSLIM BHATIYARA
MUSLIM BISWAS
MUSLIM CHUTOR MISTRI
MUSLIM DAFADAR
MUSLIM DARJI
MUSLIM DHOBI
MUSLIM GAWLI
MUSLIM GUJAR
MUSLIM HALDAR
MUSLIM HALWAI
MUSLIM JAMADAR
MUSLIM KALANDAR
MUSLIM KAYASTHA
MUSLIM LASKAR
MUSLIM LOHAR
MUSLIM MALI
MUSLIM MANYAR
MUSLIM MOLLA
MUSLIM NAI
MUSLIM NEHARIYA
MUSLIM PENCHI
MUSLIM PIYADA
MUSLIM RAJMISTRI
MUSLIM RAJPUT
MUSLIM RELIGION BHANGI
MUSLIM SALMANI
MUSLIM SANPUI
MUSLIM SARDAR
MUSLIM TELI
NADAF
OSTAGAR
PAHADIA MUSLIM
PATHAN
PATNI MUSLIM
PEERAHI
PINDARI
PINJARA
PINJARA PINJARI
PINJARI
QASSAB
QASSAB-QURESHI
QASSAI
QUASAB
QURAISHI
QURESHI
RABBAI
RANGHAR
RANGREZ
RAVUTHUR
RAWTHER
RIZVI
SAAPWALE
SAI
SAIFI
SAIN
SAIYED
SAPUI
SAYED
SAYYAD
SHAIK
SHEIKH
SHEIKH SARVARI
SHEIKH-ABBASI
SHEIKHPATHAN
SHER
SIDDIQUI
SIHAMUSLIM
SULEMANI
SUNNI
TELI MALIK
TURKATA
VAJIR
"""
# Replace this with your complete Muslim caste list (like earlier)
muslim_castes_set = set(name.strip().upper() for name in muslim_castes_raw.strip().splitlines())

# Split subcastes into Hindu and Muslim
# Split subcastes into Hindu and Muslim (as comma-separated list-like strings)
def split_subcastes(subcastes):
    hindu_caste = []
    muslim_caste = []
    for sub in subcastes:
        sub_clean = sub.strip()
        if sub_clean.upper() in muslim_castes_set:
            muslim_caste.append(sub_clean)
        else:
            hindu_caste.append(sub_clean)
    return pd.Series([", ".join(hindu_caste), ", ".join(muslim_caste)])

df[['hindu_caste', 'muslim_caste']] = df['subcastes'].apply(split_subcastes)

# Expand into separate rows
expanded_rows = []
for _, row in df.iterrows():
    caste_name = row['Caste']
    source = row['source']
    if row['hindu_caste']:
        expanded_rows.append({
            "Caste": f"{caste_name} (Hindu)",
            "Subcastes": row['hindu_caste'],
            "source": source
        })
    if row['muslim_caste']:
        expanded_rows.append({
            "Caste": f"{caste_name} (Muslim)",
            "Subcastes": row['muslim_caste'],
            "source": source
        })

df_expanded = pd.DataFrame(expanded_rows)

# ✅ Filter only Muslim rows (if needed)
df_muslim_only = df_expanded[df_expanded["Caste"].str.endswith(" (Muslim)")].reset_index(drop=True)
df_hindu = df_expanded[df_expanded["Caste"].str.endswith(" (Hindu)")].reset_index(drop=True)


# Optional: save result
# df_muslim_only.to_csv("muslim_only_castes.csv", index=False)

# View first few rows
df_muslim_only
```
