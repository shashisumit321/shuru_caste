```python
import pandas as pd
import ast 
```

```python
df = pd.read_csv("Caste Analysis and QC - unique_caste_for_every_state (1).csv")
```

```python
df
```

```python
df['subcastes'] = df['subcastes'].apply(ast.literal_eval)
```

```python
unique_caste =  (
    df.groupby('Caste')['subcastes']
    .apply(lambda lists: list(set([item for sublist in lists for item in sublist])))
    .reset_index()
)

unique_caste
```

```python
unique_caste
```

```python
unique_caste.to_csv("final_unique_caste.csv", index=False)
```
